```python
!pip install chardet
import chardet

def detect_encoding(filename):
    with open(filename, 'rb') as f:
        bytes = f.read()
    return chardet.detect(bytes)['encoding']
```

```python
import pandas as pd
```

```python
prouni_2016 = pd.read_csv('/content/drive/MyDrive/Prouni/pda-prouni-2016.csv', sep=';', encoding=detect_encoding('/content/drive/MyDrive/Prouni/pda-prouni-2016.csv'))
prouni_2017 = pd.read_csv('/content/drive/MyDrive/Prouni/pda-prouni-2017.csv', sep=';', encoding=detect_encoding('/content/drive/MyDrive/Prouni/pda-prouni-2017.csv'))
prouni_2018 = pd.read_csv('/content/drive/MyDrive/Prouni/pda-prouni-2018.csv', sep=';', encoding=detect_encoding('/content/drive/MyDrive/Prouni/pda-prouni-2018.csv'))
prouni_2019 = pd.read_csv('/content/drive/MyDrive/Prouni/pda-prouni-2019.csv', sep=';', encoding=detect_encoding('/content/drive/MyDrive/Prouni/pda-prouni-2019.csv'))
prouni_2020 = pd.read_csv('/content/drive/MyDrive/Prouni/pda-prouni-2020.csv', sep=';', encoding=detect_encoding('/content/drive/MyDrive/Prouni/pda-prouni-2020.csv'))
```

#**CONCATENAR (JUNTAR CONSULTAS)**

```python
# ignore_index = True - ignora o index automatico que o python coloca
# keys (quando utiliza esse não pode utilizar o ignore) - ele traz uma coluna especificando de qual arquivo é

df = pd.concat([prouni_2016, prouni_2017, prouni_2018, prouni_2019, prouni_2020], keys=['2016', '2017', '2018', '2019', '2020'])
```

```python
df.to_csv('/content/drive/MyDrive/Prouni/prouni_unificado_tratado.csv')
```
