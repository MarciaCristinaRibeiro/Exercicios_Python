#**Estatística Descritiva**

* As estatísticas descritivas envolvem os métodos usados para resumir, simplificar e descrever os principais recursos de um conjunto de dados.

* Este ramo da estatística visa fornecer uma visão geral concisa e significativa dos dados por meio de medidas como tendência central (média, mediana, moda), dispersão (intervalos, variância, desvio-padrão) e representações gráficas, tais como histogramas, gráficos de barras, gráfico de setores, gráficos de dispersão e etc.

```python
# Existe alguma relação entre curtidas e comentarios
plt.figure(figsize=(10,3))
plt.scatter(df['qtd_curtidas'], df['qtd_comentarios'], color='black')
plt.xlabel('Quantidade de Curtidas')
plt.ylabel('Quantidade de Comentários')
plt.title('Curtidas vs. Comentários')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Dados (substitua isso pelos seus próprios dados)
qtd_curtidas = df['qtd_curtidas']
qtd_comentarios = df['qtd_comentarios']

# Calculando a linha de regressão e o R²
slope, intercept, r_value, p_value, std_err = stats.linregress(qtd_curtidas, qtd_comentarios)

# Plotando o gráfico de dispersão
plt.figure(figsize=(10,6))
plt.scatter(qtd_curtidas, qtd_comentarios, color='black', label='Dados')

# Adicionando a linha de regressão
plt.plot(qtd_curtidas, intercept + slope * qtd_curtidas, color='red', label='Linha de Regressão')

# Adicionando a equação da linha de regressão
equacao_regressao = f'Regressão: y = {slope:.2f}x + {intercept:.2f}'
plt.text(0.5, 0.8, equacao_regressao, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=12, color='blue')

# Adicionando o valor de R² no gráfico
plt.text(0.5, 0.9, f'R² = {r_value**2:.2f}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=12)

plt.xlabel('Quantidade de Curtidas')
plt.ylabel('Quantidade de Comentários')
plt.title('Curtidas vs. Comentários')
plt.legend()
plt.show()
```
